```python
import pandas as pd

def get_df(code, start, end):
    df = pd.read_csv('/data/s/his_qfq_%s.csv' % code)
    df = df[ df.date >= start ]
    df = df[ df.date <= end ]
    
#     return df
    
    df.index = pd.DatetimeIndex(df.date)
    return df['close']
```

```python
names = ['601600','600222']
df = pd.DataFrame({c:get_df(c,'2016-01-01','2016-12-31') for c in names } ,columns=names)
```

```python
df = df.asfreq('B').fillna(method='pad')
```

```python
rets = df.pct_change()
rets

rets = ((1 + df.pct_change()).cumprod() - 1)
# rets[:1] = 0



t = rets.T
t.apply(len)
```

```python
def calc_mom(price, lookback, lag):
    mom_ret = price.shift(lag).pct_change(lookback)
    ranks = mom_ret.rank(axis=1, ascending=False)
    demeaned = ranks - ranks.mean(axis=1)
    return demeaned / demeaned.std(axis=1)
```
