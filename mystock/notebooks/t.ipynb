```python
import requests
import pandas as pd
import json
import time

def fetch_data(url, file_name):
    dfs = []
    s = requests.Session()
    for i in range(1,40):
        while True:
            try:
                r = s.get(url.replace('*****',str(i)))
                t = r.text[28:-1]
#                 print(t)
#                 print(t.text)
                print(i, t[:50])

                j = json.loads(t)
                df = pd.DataFrame.from_dict(j)
                dfs.append(df)
                time.sleep(2)
                break
            except:
                s = requests.Session()
#                 break
                time.sleep(60)
#         break
        

    all_df = pd.concat(dfs)
    all_df.index = pd.DatetimeIndex(all_df.tdate)
    try:
        all_df.to_excel(file_name)
    except:
        print('error to create file...')
    return all_df
```
